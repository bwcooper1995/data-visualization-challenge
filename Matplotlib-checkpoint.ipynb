```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
mouse_metadata_path = "Starter_Code/Pymaceuticals/data/Mouse_metadata.csv"
study_results_path = "Starter_Code/Pymaceuticals/data/Study_results.csv"
```

```python
mouse_meta = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)
combined_mouse_study = pd.merge(study_results, mouse_meta, on="Mouse ID")
combined_mouse_study
```

```python
number_of_mice = combined_mouse_study["Mouse ID"].unique()
len(number_of_mice)
```

```python
duplicate_check = combined_mouse_study[combined_mouse_study.duplicated(['Mouse ID', 'Timepoint'], keep=False)]
duplicate_check
```

```python
cleaned_data = combined_mouse_study.loc[combined_mouse_study["Mouse ID"].isin(["g989"]) == False]

cleaned_data
```

```python
cleaned_uniques = cleaned_data["Mouse ID"].unique()
len(cleaned_uniques)
```

```python
mean_tumor = cleaned_data.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
median_tumor = cleaned_data.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
var_tumor = cleaned_data.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
std_tumor = cleaned_data.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
sem_tumor = cleaned_data.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]
```

```python
tumors = {"Mean Tumor Volume" : mean_tumor,
          "Median Tumor Volume" : median_tumor,
          "Tumor Volume Variance" : var_tumor,
          "Tumor Volume Std. Dev." : std_tumor,
          "Tumor Volume Std. Err." : sem_tumor}

tumor_summary = pd.DataFrame(tumors)
```
